```python
import glob
import numpy as np
import cv2


for image in glob.glob("/Users/andrewcross/Desktop/MHFDDrone Data/Mannings_20210505/Mannings_20210505/*.png"):
    im = cv2.imread(image)
    height, width, channels = np.shape(im)
    for i in range(0,height):
        for j in range(0,width):
            for p in range(0,channels):
                if im[i,j,p] == 7 or im[i,j,p] ==13:
                    im[i,j,p] = 1
                elif im[i,j,p] == 8 or im[i,j,p] ==14:
                    im[i,j,p]=2
                elif im[i,j,p]==9 or im[i,j,p] ==15 :
                    im[i,j,p]=3
                elif im[i,j,p]==10 or im[i,j,p] ==16 :
                    im[i,j,p]=4
                elif im[i,j,p]==11 or im[i,j,p] ==17:
                    im[i,j,p]=5
                elif im[i,j,p]==12 or im[i,j,p] ==18 or im[i,j,p] == 6:
                    im[i,j,p]=4
    name=image.split("/")
    name=name[-1]
    print(name)
    cv2.imwrite("/Users/andrewcross/Desktop/MHFDDrone Data/Mannings_20210505/Mannings_20210505/mannings_edited/"+str(name),im)
```
